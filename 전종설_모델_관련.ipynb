```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, StackingClassifier
import xgboost as xgb
from sklearn.metrics import accuracy_score

# Create individual models
estimators = [
    ('decision_tree', DecisionTreeClassifier()),
    ('random_forest', RandomForestClassifier(n_estimators = 100, oob_score=True, random_state= 1)),
    ('xgboost', XGBClassifier(n_estimators=1000, learning_rate=0.2, max_depth=10, random_state = 1))
]

# Create the Stacking ensemble model
stacking_model = StackingClassifier(estimators=estimators, final_estimator=DecisionTreeClassifier())

# Train the stacking model
stacking_model.fit(x_train, y_train)
```

```python
import math
import time

start = time.time()
# Make predictions on the test set
stacking_predictions = stacking_model.predict(x_test)
end = time.time()

print(f"{end - start:.5f} sec")
# Calculate the accuracy of the stacking ensemble
stacking_accuracy = accuracy_score(y_test, stacking_predictions)
print("Stacking Ensemble Accuracy:", stacking_accuracy)
```

```python
from sklearn.ensemble import VotingClassifier
from sklearn.tree import DecisionTreeClassifier
from xgboost import XGBClassifier



# Create individual models
models = [
    ('decision_tree', DecisionTreeClassifier()),
    ('random_forest', RandomForestClassifier(n_estimators = 100, oob_score=True, random_state= 1)),
    ('xgboost', XGBClassifier(n_estimators=1000, learning_rate=0.2, max_depth=10, random_state = 1))
]

# Create the Voting ensemble model
voting_model = VotingClassifier(estimators=models)

# Train the voting model
voting_model.fit(x_train, y_train)

# Make predictions on the test set
voting_predictions = voting_model.predict(x_test)

# Calculate the accuracy of the voting ensemble
voting_accuracy = accuracy_score(y_test, voting_predictions)
print("Voting Ensemble Accuracy:", voting_accuracy)
```
